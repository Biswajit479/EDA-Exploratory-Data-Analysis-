```python
# In the dataset all the column is object.But 'actual_price','discount_percentage','rating','rating_count' both are numeric type but here is object. so change the datatype of the columns
```

```python
# df['actual_price(₹)'] = df['actual_price'].map({'₹':'',',':''})
df['actual_price(₹)'] = df['actual_price'].str.replace('₹','')
df['actual_price(₹)'] = df['actual_price(₹)'].str.replace(',','')
df['actual_price(₹)']
```

```python
# so now dop the 'actual_price' column
df.drop('actual_price', axis = 1, inplace = True)
```

```python
# now change the datatype of 'actual_price(₹)' column
df['actual_price(₹)'] = df['actual_price(₹)'].astype(float)
df['actual_price(₹)']
```

```python
df['discount_percentage'].unique()
```

```python
# in 'discount_percentage' column replace '%' with empty string
df['discount_percentage'] = df['discount_percentage'].str.replace('%','')

# change the datatype of the column
df['discount_percentage'] = df['discount_percentage'].astype(float)
df['discount_percentage'].dtype
```

```python
# rating
df['rating'].unique()
```

```python
# # we see that 'rating' column contains '|' that value replace
df['rating'] = df['rating'].str.replace('|','', regex = False)
# now convert '' to None value
df['rating'] = df['rating'].replace('',np.nan)
# # now convert into float
df['rating'] = df['rating'].astype(float)
# # check the datatype
df['rating'].dtype
```

```python
# check the unique value of rating column
df['rating'].unique()
```

```python
# now fill the null value intoo mean value in the rating column
df['rating'].fillna(round(df['rating'].mean(),2), inplace = True)
```

```python
# check 'rating_count' column
df['rating_count'].unique()
```

```python
# in 'rating_count' column contains ',' replace with '' empty space
df['rating_count'] = df['rating_count'].str.replace(',','')
# now change the datatype
df['rating_count'] = df['rating_count'].astype(float)
# check the datatype
df['rating_count'].dtype
```

```python
# now check the datatype of the columns
df.dtypes
```

```python
df
```

```python
# Similarly actual_price column we treat the discount_price
df['discount_price(₹)'] = df['discounted_price'].str.replace('₹','')
df['discount_price(₹)'] = df['discount_price(₹)'].str.replace(',','')

# drop the original discount_price
df.drop('discounted_price', axis = 1, inplace = True)

# now change the datatype of discount_price(₹) column
df['discount_price(₹)'] = df['discount_price(₹)'].astype(float)

# check the dataframe
df
```

```python
# check the null value present of each column
df.isnull().sum()
```

```python
# rating_count column 2 null value so, fill the null value with mean value
mean_value = round(df['rating_count'].mean(),2)
df['rating_count'] = df['rating_count'].fillna(mean_value)
```

```python
# now check the null value
df.isnull().sum()
```

# Questions:

## 1. What is the average rating for each product category?

```python
avg_rating_product = df.groupby('product_name')['rating'].mean()
print("Average rating for each product Category:\n",avg_rating_product)
```

## 2. What are the top rating_count products by category?

```python
top_rating_count_category = df.groupby('category')['rating_count'].std().sort_values(ascending = False)
print("Top 10 Rating count by category:\n",top_rating_count_category.head(10))
```

## 3. What is the distribution of discounted prices vs. actual prices?

```python
distribution_discount_actual_price = df[['discount_price(₹)','actual_price(₹)']].describe().T
print("Distribution of Discounted and Actual Price:\n",distribution_discount_actual_price[['min','25%','50%','75%','max']])
```

## 4.How does the average discount percentage vary across categories?

```python
avg_discount_percentage_categories = df.groupby('category')['discount_percentage'].mean()
print("The Average discount percentage vary across categories:\n",avg_discount_percentage_categories)
```

## 5. What are the most popular product names?

```python
popular_product_name = df['product_name'].value_counts(dropna = True).sort_values(ascending = False)
print("Top 10 Product name:\n",popular_product_name.head(10))
```

## 6. What are the most popular product keywords?

```python
top_10_popular_product_keywords = df['product_id'].value_counts().sort_values(ascending = False).head(10)
print("Top 10 popular product keyword:\n",top_10_popular_product_keywords)
```

## 7.What are the most popular product reviews?

```python
# Step 1: Create a new column 'review_count' using transform to align with the original DataFrame
df['review_count'] = df.groupby('product_name')['review_title'].transform('count')

# Step 2: Sort and remove duplicates to get one row per product
most_reviewed_product = df[['product_name', 'category', 'review_title', 'rating', 'review_count']].drop_duplicates('product_name').sort_values(by='review_count', ascending=False).reset_index(drop=True)

# Step 3: Print the top results
print("Most Reviewed product:\n", most_reviewed_product.head(10))
```

## 8. What is the correlation between discounted_price and rating?

```python
corr = df[['discount_price(₹)','rating']].corr()
corr
```

## 9.  What are the Top 5 categories based on the highest ratings?

```python
top_5_category_based_highest_rating = df[['category','rating']].drop_duplicates('category').sort_values(by = 'rating', ascending = False).reset_index(drop =True).head()
print("Top 5 Categories based on highest ratings:\n",top_5_category_based_highest_rating)
```
