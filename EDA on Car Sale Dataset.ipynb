# Questions:

## 1. What is the average selling price of cars for each dealer, and how does it compare across different dealers?

```python
dealer_avg_price = round(df.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending = False),3)
dealer_avg_price
```

```python
top_10_dealer_avg_price = dealer_avg_price.head(10)
top_10_dealer_avg_price.plot(kind = 'bar', figsize = (20,10), color = 'skyblue')
plt.title("Top 10 Dealers by Average Selling Price")
plt.xlabel("Average Selling Selling Price ($)")
# plt.show()

for i,v in enumerate(top_10_dealer_avg_price):
    plt.text(i, v+5, str(v), ha = 'center', va = 'bottom', fontsize = 10)
```

## 2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends.

```python
company_price_variation = round(df.groupby('Company')['Price ($)'].std().sort_values(ascending = False),3)
company_price_variation
```

```python
top_10_company_price_variation = company_price_variation.head(10)
top_10_company_price_variation.plot(kind = 'bar', figsize = (12,10), color = 'green')

for i,v in enumerate(top_10_company_price_variation):
    plt.text(i, v+5, str(v), ha = 'center', va = 'bottom', fontsize = 10)
```

## 3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

```python
# group by transmission and describe the price distribution
transmission_stats = df.groupby('Transmission')['Price ($)'].describe()

# calculate IQR = Q3-Q1
transmission_stats['IQR'] = transmission_stats['75%'] - transmission_stats['25%']

# Show relevant summary including IQR
print(transmission_stats[["min","25%","50%","75%","max","IQR"]])
```

## 6. How does the average selling price of cars vary by customer gender and annual income?

```python
# create a income bins for cateegorization
income_bins = [0, 50000, 100000, 250000, 500000, 1000000, float('inf')]
income_labels = ['<50K', '50K–100K', '100K–250K', '250K–500K', '500K–1M', '>1M']

df['Income Group'] = pd.cut(df['Annual Income'], bins = income_bins, labels = income_labels)

average_price_by_gender_income = df.groupby(['Gender','Income Group'])['Price ($)'].mean().unstack()
print(average_price_by_gender_income)
```

## 7. What is the distribution of car prices by region, and how does the number of cars sold vary by region?

```python
price_distribution_by_region = df.groupby('Dealer_Region')['Price ($)'].describe()

# Count number of car sold per region
cars_sold_by_region = df['Dealer_Region'].value_counts()

print("Price Distribution by region:\n", price_distribution_by_region[["min", "25%", "50%", "75%", "max"]])
print("\nCars sold by region:\n",cars_sold_by_region)
```

```python
plt.figure(figsize = (12,10))
sns.boxplot(x = 'Dealer_Region', y = 'Price ($)', data = df, palette = 'Set2')
plt.tight_layout()
plt.title('Price Distribution by region')
plt.show()
```

```python
cars_sold_by_region.plot(kind = 'bar', color = 'yellow')
plt.title('Cars Sold by region')
plt.xticks(rotation = 60)

for i,v in enumerate(cars_sold_by_region):
    plt.text(i, v+5, str(v), ha = 'center', va = 'bottom', fontsize = 10)
```

## 10. What are the top 5 car models with the highest number of sales, and how does their price distribution look?

## 14. What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?

```python
import pandas as pd

# Load data
df = pd.read_csv("Car Sale.csv")
df.columns = df.columns.str.strip()

# Try to extract engine size as numeric (if present)
df['Engine_Size'] = df['Engine'].str.extract(r'(\d+\.?\d*)').astype(float)

# Select numeric columns for correlation
corr_data = df[['Price ($)', 'Annual Income', 'Engine_Size']]

# Drop missing values
corr_data = corr_data.dropna()

# Calculate correlation matrix
correlation_matrix = corr_data.corr()

print("🔍 Correlation Matrix:\n", correlation_matrix)
```
