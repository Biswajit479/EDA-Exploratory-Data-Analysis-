## 2. What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.

```python
plt.figure(figsize = (12,10))
plt.hist(df['Popularity'].dropna(), bins = 10, edgecolor = 'black', color = 'skyblue')

plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity Score')
plt.ylabel('Number of Tracks')

plt.tight_layout()

plt.show()
```

## 3. Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

```python
plt.figure(figsize = (12,10))

sns.scatterplot(x = 'Popularity', y = 'Duration (ms)', data = df, palette = 'dark')

plt.title('Relationship between popularity and duration of tracks')
plt.xlabel('Popularity of the Track')
plt.ylabel('Duration time of the track in ms')

plt.show()
```

## 4. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.

```python
artist_counts = df['Artist'].value_counts().reset_index()

top_artist = artist_counts.iloc[0]

print(f"'{top_artist['Artist']}' has the highest number of tracks and the number of tracks is  = {top_artist['count']}")
```

```python
plt.figure(figsize = (12,10))
sns.countplot(x = 'Artist', data = df, order = df['Artist'].value_counts().index[:10], palette = 'coolwarm')

plt.title("Top 10 Artists by Number of Tracks")
plt.xlabel('Artist name')
plt.ylabel('Number of Tracks')

plt.tight_layout()
plt.show()
```

## 5. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.

## 6. Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist.

## 9. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.

```python
top_artist = df['Artist'].value_counts().head(10).index
filter_df = df[df['Artist'].isin(top_artist)]

plt.figure(figsize = (12,10))

plt.subplot(1,2,1)
sns.boxplot(x = 'Artist', y = 'Duration (ms)', data = filter_df, palette = 'dark')
plt.title('Track Duration by Artist (Box Plot)')
plt.xticks(rotation = 60)

plt.subplot(1,2,2)
sns.violinplot(x = 'Artist', y = 'Duration (ms)', data = filter_df, palette = 'Set3')
plt.title('Track Duration by Artist (Violin Plot)')
plt.xticks(rotation = 60)

plt.show()
```

## 10.  How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.

```python
top_artist = df['Artist'].value_counts().head(10).index
filter_df = df[df['Artist'].isin(top_artist)]

plt.figure(figsize = (12,10))

plt.subplot(1,2,1)
sns.violinplot(x = 'Artist', y = 'Popularity', data = filter_df, palette = 'dark')
plt.title('Track Duration by Artist (Violin Plot)')
plt.xticks(rotation = 60)

plt.subplot(1,2,2)
sns.swarmplot(x = 'Artist', y = 'Popularity', data = filter_df, palette = 'Set3')
plt.title('Track Duration by Artist (swarm Plot)')
plt.xticks(rotation = 60)

plt.show()
```
